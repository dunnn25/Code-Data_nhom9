```python
import pandas as pd
import matplotlib.pyplot as plt
from neuralprophet import NeuralProphet
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Đọc dữ liệu từ file CSV
df = pd.read_csv('/content/Hashtag_sum_2_10_2013.csv')
def parse_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%d/%m/%Y')
    except:
        return pd.to_datetime(date_str, format='%m/%d/%Y')

df['date'] = df['date'].apply(parse_date)


# Điền giá trị thiếu bằng nội suy tuyến tính
df = df.set_index('date').reindex(pd.date_range(start=df['date'].min(), end=df['date'].max(), freq='D')).interpolate(method='linear').reset_index()
df = df.rename(columns={'index': 'date'})

# Chuẩn bị dữ liệu cho NeuralProphet
df_negative = df[['date', '#covid19']].rename(columns={'date': 'ds', '#covid19': 'y'})
df_negative['ds'] = pd.to_datetime(df_negative['ds'])

# Chuẩn hóa dữ liệu
scaler = MinMaxScaler()
# Fit the scaler on the 'y' column before using inverse_transform
df_negative['y'] = scaler.fit_transform(df_negative[['y']].fillna(0))


# Huấn luyện lại mô hình (nếu cần, hoặc tải mô hình đã huấn luyện)
model = NeuralProphet(
    n_lags=5,
    n_forecasts=1,  # Dự đoán 1 ngày tại mỗi thời điểm
    n_changepoints=100,
    yearly_seasonality=False,
    weekly_seasonality=False,
    daily_seasonality=True,
    batch_size=64,
    epochs=50,
    learning_rate=0.005
)

# Thêm biến ngoại sinh
# model = model.add_lagged_regressor(names=['Neutral', 'Positive'])

# Huấn luyện mô hình
model.fit(df_negative, freq='D')

# Dự đoán trên tập huấn luyện
forecast = model.predict(df_negative)

# Đảo ngược chuẩn hóa
forecast['yhat1'] = scaler.inverse_transform(forecast[['yhat1']].fillna(0))
# The y column in df_negative is already scaled, so we only need to inverse transform forecast['yhat1']
# df_negative['y'] = scaler.inverse_transform(df_negative[['y']].fillna(0))

# Tạo biểu đồ đường
plt.figure(figsize=(12, 6))

# Plot the actual values
plt.plot(df['date'], df['#covid19'], label='Covid 19 (Actual)', color='#FF6B6B', linewidth=2)

# Plot the predicted values from the forecast dataframe
plt.plot(forecast['ds'], forecast['yhat1'], label='Covid 19 (Predicted)', color='#FFA500', linestyle='--', linewidth=2)

# Add title and labels
plt.title('Comparison of Actual vs Predicted on Training Data', fontsize=14, pad=15)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Counts', fontsize=12)

# Add legend
plt.legend()

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Customize x-axis date format
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.AutoDateLocator())
plt.xticks(rotation=45) # Rotate dates for better visibility

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

# For evaluation, we need the unscaled true values.
# Since we scaled df_negative['y'] in place, let's reload the original df_negative or use the original df column
# For evaluation, it's better to use the original unscaled data for y_true
y_true = df['#covid19'][forecast.index] # Align y_true with forecast index
y_pred = forecast['yhat1']


# Drop rows with NaN in either y_true or y_pred for accurate calculation
valid_indices = ~y_true.isna() & ~y_pred.isna()
y_true_aligned = y_true[valid_indices]
y_pred_aligned = y_pred[valid_indices]


# Tính toán các chỉ số đánh giá
mae = mean_absolute_error(y_true_aligned, y_pred_aligned)
mse = mean_squared_error(y_true_aligned, y_pred_aligned)
rmse = np.sqrt(mse) # RMSE là căn bậc hai của MSE
r2 = r2_score(y_true_aligned, y_pred_aligned)

# In các chỉ số đánh giá
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
print(f'R-squared (R2): {r2:.2f}')
```
