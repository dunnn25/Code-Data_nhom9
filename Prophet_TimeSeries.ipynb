```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
from datetime import datetime

# Load the dataset
df = pd.read_csv('/content/Hashtag_sum_2_10_2013.csv')

# Convert date column to datetime, handling mixed formats
def parse_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%d/%m/%Y')
    except:
        return pd.to_datetime(date_str, format='%m/%d/%Y')

df['date'] = df['date'].apply(parse_date)

# Filter for the last 5 years (April 1, 2018, to April 20, 2023)
start_date = '2018-04-01'
end_date = '2023-04-20'
df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Create a complete date range and fill missing dates
all_dates = pd.date_range(start=start_date, end=end_date, freq='D')
df_full = pd.DataFrame({'date': all_dates})
df = df_full.merge(df, on='date', how='left').fillna(0)

# List of hashtags to forecast
hashtags = ['#covid19', '#ukraine', '#travel', '#mondaymotivation', '#christmas']

# Function to forecast for a given hashtag
def forecast_hashtag(hashtag, df):
    # Prepare data for Prophet
    df_prophet = df[['date', hashtag]].rename(columns={'date': 'ds', hashtag: 'y'})

    # Initialize and fit the Prophet model
    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=True,
        daily_seasonality=False,
        changepoint_prior_scale=0.05   # Adjust for flexibility in trend changes
    )
    model.fit(df_prophet)

    # Create future dataframe for 30 days
    future = model.make_future_dataframe(periods=30)

    # Forecast
    forecast = model.predict(future)

    # Plot historical data and forecast as line plots
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(df_prophet['ds'], df_prophet['y'], label='Historical Data')
    ax.plot(forecast['ds'], forecast['yhat'], label='Forecast', color='red')
    ax.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='red', alpha=0.2)
    ax.set_title(f'Forecast for {hashtag}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Hashtag Count')
    ax.legend()
    plt.grid(True)
    plt.show()

    # Plot components (trend, yearly, weekly)
    fig2 = model.plot_components(forecast)
    plt.show()

    # Return forecast for the next 30 days
    return forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(30)

# Run forecasting for each hashtag
forecasts = {}
for hashtag in hashtags:
    print(f"\nForecasting for {hashtag}")
    # Truyền bản sao của DataFrame để tránh các vấn đề sửa đổi tiềm ẩn
    forecasts_hashtag = forecast_hashtag(hashtag, df.copy())
    if forecasts_hashtag is not None:
        forecasts.update({hashtag: forecasts_hashtag})

# Print forecast results
for hashtag, forecast in forecasts.items():
    print(f"\n30-Day Forecast for {hashtag}:")
    print(forecast[['ds', 'yhat']].round(0))
```

```python
# prompt: đánh giá mô hình theo mse, rmse, r2, mae

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Assuming you have actual values for the forecast period to compare against
# Let's create a placeholder for actual values for demonstration
# In a real scenario, you would load or define your actual values for the 30-day forecast period

# For demonstration, let's extract the historical 'y' values to simulate actuals for the end of the historical period
# and the beginning of the forecast period. In a real scenario, you would have a separate test set
# with actual values for the forecast dates.

evaluation_results = {}

for hashtag in hashtags:
    print(f"\nEvaluating model for {hashtag}")

    # Get the historical data used for training
    df_prophet_historical = df[['date', hashtag]].rename(columns={'date': 'ds', hashtag: 'y'})
    df_prophet_historical = df_prophet_historical[df_prophet_historical['ds'] <= end_date]

    # Get the forecast for the historical period
    # We need to re-run the prediction on the historical dates to get yhat for evaluation
    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=True,
        daily_seasonality=False,
        changepoint_prior_scale=0.05
    )
    model.fit(df_prophet_historical)
    forecast_historical = model.predict(df_prophet_historical[['ds']])

    # Merge historical actuals and historical predictions
    evaluation_df = df_prophet_historical.merge(forecast_historical[['ds', 'yhat']], on='ds', how='left')

    # Drop rows where 'y' is NaN (if any, although filling NaNs with 0 earlier should prevent this)
    evaluation_df = evaluation_df.dropna(subset=['y'])

    # Calculate evaluation metrics on the historical data (as a proxy for evaluation)
    # In a real scenario, you would evaluate on a separate hold-out test set
    actual_values = evaluation_df['y']
    predicted_values = evaluation_df['yhat']

    if len(actual_values) > 0:
        mse = mean_squared_error(actual_values, predicted_values)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(actual_values, predicted_values)
        r2 = r2_score(actual_values, predicted_values)

        evaluation_results[hashtag] = {
            'MSE': mse,
            'RMSE': rmse,
            'MAE': mae,
            'R2': r2
        }
        print(f"  MSE: {mse:.4f}")
        print(f"  RMSE: {rmse:.4f}")
        print(f"  MAE: {mae:.4f}")
        print(f"  R2: {r2:.4f}")
    else:
        print(f"  No historical data available for evaluation for {hashtag}.")

print("\nSummary of Evaluation Results:")
for hashtag, metrics in evaluation_results.items():
    print(f"\nHashtag: {hashtag}")
    for metric, value in metrics.items():
        print(f"  {metric}: {value:.4f}")

```
