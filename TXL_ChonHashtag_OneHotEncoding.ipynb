```python
df = pd.read_csv('/content/hashtag_only_8_p2.csv') # Chọn các file cần tiền xử lý ví dụ lấy 3 file 8 9 10
df_4 = pd.read_csv('/content/hashtag_only_9_p2.csv')
df_3 = pd.read_csv('/content/hashtag_only_10_p2.csv')
```

```python
# Đếm số lượng hashtag xuất hiện trong cột hashtag chỉ lấy kí tự trong khoảng dấu ' '

from collections import Counter
import re

hashtags = df['hashtag'].tolist()
hashtag_counts = Counter()

for tag_string in hashtags:
    # Find all hashtags within the string (words starting with #)
    found_hashtags = re.findall(r'#\w+', tag_string)
    for tag in found_hashtags:
        hashtag_counts[tag] += 1

# Print the counts
for tag, count in hashtag_counts.most_common():
    print(f"{tag}: {count}")
```

```python
# gộp các hashtag trùng nhau và đếm xong sắp xếp giảm dần

import pandas as pd
# Combine hashtags and count them
hashtag_counts = Counter()

for tag_string in hashtags:
    # Find all hashtags within the string (words starting with #)
    # The regex r'#\w+' correctly identifies hashtags like #example
    found_hashtags = re.findall(r'#\w+', str(tag_string)) # Ensure tag_string is a string
    for tag in found_hashtags:
        hashtag_counts[tag.lower()] += 1 # Convert to lowercase to treat #Tag and #tag as the same

# Create a DataFrame from the counts
df_hashtag_counts = pd.DataFrame.from_dict(hashtag_counts, orient='index', columns=['count'])

# Sort the DataFrame by count in descending order
df_hashtag_counts = df_hashtag_counts.sort_values(by='count', ascending=False)

# Reset index to make the hashtag a column
df_hashtag_counts = df_hashtag_counts.reset_index()
df_hashtag_counts.rename(columns={'index': 'hashtag'}, inplace=True)

# Print the resulting DataFrame
df_hashtag_counts.head(50)
```

```python
# giữ lại các dòng dữ liệu có 1 trong 10 hashtag sau: #covid19, #crypto, #bitcoin, #ai, #ukraine, #travel, #mondaymotivation, #christmas, #iran, #realestate cho df
import re
hashtags_to_keep = ['#covid19', '#covid', 'coronavirus', '#crypto', '#bitcoin', '#ukraine', '#travel', '#mondaymotivation', '#christmas', '#iran', '#realestate']

# Create a boolean mask where each row is True if it contains at least one of the hashtags
mask = df['hashtag'].str.contains('|'.join(hashtags_to_keep), case=False, na=False)

# Filter the DataFrame
df_filtered = df[mask].copy()

# Display the first few rows of the filtered DataFrame
print(df_filtered.head())

# Display the number of rows in the original and filtered DataFrames
print(f"\nOriginal DataFrame shape: {df.shape}")
print(f"Filtered DataFrame shape: {df_filtered.shape}")
```

Tiến hành tạo cột và one-hot encoding

```python
# tạo thêm 9 cột cho 9 hashtag: #covid19, #crypto, #bitcoin #ukraine, #travel, #mondaymotivation, #christmas, #iran, #realestate, nếu hàng nào xuất hiện các hashtag như cột thì cho dữ liệu của cột và hàng đó là 1, không xuất hiện thì cho 0, riêng các hashtag #covid, #coronavirus thì cho chung vào cột covid19 (nếu xuất hiện thì vẫn chỉ cho 1)

# Define the list of target hashtags and their corresponding column names
target_hashtags = {
    '#covid19': ['#covid19', '#covid', '#coronavirus'],
    '#crypto': ['#crypto'],
    '#bitcoin': ['#bitcoin'],
    '#ukraine': ['#ukraine'],
    '#travel': ['#travel'],
    '#mondaymotivation': ['#mondaymotivation'],
    '#christmas': ['#christmas'],
    '#iran': ['#iran'],
    '#realestate': ['#realestate']
}

# Create new columns and initialize with 0
for col_name in target_hashtags.keys():
    df_filtered[col_name] = 0

# Iterate through each row and update the new columns
for index, row in df_filtered.iterrows():
    hashtag_string = str(row['hashtag']).lower() # Ensure it's a string and lowercase

    for col_name, hashtags in target_hashtags.items():
        for hashtag in hashtags:
            if hashtag in hashtag_string:
                df_filtered.loc[index, col_name] = 1
                break # If any of the associated hashtags are found, set to 1 and move to the next column

# Display the first few rows to verify the new columns
print(df_filtered.head())
```

```python
# prompt: chuyển cột date về thành dạng ngày/tháng/năm và sắp xếp theo thứ tự tăng dần cho bộ dữ liệu /content/Hashtag_encoded_p2.csv

import pandas as pd
# Load the dataset
df_encoded = pd.read_csv('/content/Hashtag_encoded_p2.csv')

try:
    df_encoded['date'] = pd.to_datetime(df_encoded['date'])
except KeyError:
    print("Error: 'date' column not found. Please check the column name.")
    # Exit or handle the error appropriately if the date column is missing
    # For now, we'll assume the column exists for the next steps.

# Format the 'date' column to 'ngày/tháng/năm' (DD/MM/YYYY)
# This converts the datetime objects back to strings with the desired format
try:
    df_encoded['date'] = df_encoded['date'].dt.strftime('%d/%m/%Y')
except KeyError:
    print("Cannot format 'date' column as it was not found or converted.")


# Sort the DataFrame by the 'date' column in ascending order
# First, ensure the column is in a sortable format (datetime objects are ideal for sorting chronologically)
# We convert it back to datetime for sorting, handling potential errors
try:
    df_encoded['date_sortable'] = pd.to_datetime(df_encoded['date'], format='%d/%m/%Y', errors='coerce')
    df_sorted = df_encoded.sort_values(by='date_sortable', ascending=True)
    # Drop the temporary sortable column if you don't need it
    df_sorted = df_sorted.drop(columns=['date_sortable'])
except KeyError:
    print("Cannot sort by 'date' column as it was not found or converted correctly.")
    # If sorting fails, just keep the original DataFrame structure after formatting the date string
    df_sorted = df_encoded


# Display the first few rows of the modified DataFrame to verify
print(df_sorted.head())

# You can also display the data types to confirm the changes
print(df_sorted.info())
```

```python
# trong df_sorted, nếu xuất hiện các ngày trùng nhau thì gộp ngày đó vào 1, các cột '#covid19', '#covid', 'coronavirus', '#crypto', '#bitcoin', '#ukraine', '#travel', '#mondaymotivation', '#christmas', '#iran', '#realestate' sẽ được cộng vào với dữ liệu tương ứng của các cột gộp đó, giữ lại các ngày không bị trùng

# List of columns to sum during aggregation
columns_to_sum = ['#covid19', '#crypto', '#bitcoin', '#ukraine', '#travel', '#mondaymotivation', '#christmas', '#iran', '#realestate']

# Group by the 'date' column and sum the specified columns
# If a date only appears once, the aggregation will just return the values for that row.
df_aggregated = df_sorted.groupby('date')[columns_to_sum].sum().reset_index()

# Print the resulting aggregated DataFrame
df_aggregated
```

```python
# gộp các ngày trùng nhau và cộng tổng dữ liệu của cột tương ứng cho file Hashtag_sum_8_9_10.csv,  giữ ngày xuất hiện 1 lần, cuối cùng sắp xếp ngày theo thứ tự tăng dần

import pandas as pd
df_sum = pd.read_csv('/content/Hashtag_sum_8_9_10.csv')

# Convert 'date' column to datetime objects for proper sorting
# Use errors='coerce' to turn invalid date formats into NaT (Not a Time)
df_sum['date'] = pd.to_datetime(df_sum['date'], format='%d/%m/%Y', errors='coerce')

# Remove rows where date conversion failed (optional, but good practice)
df_sum.dropna(subset=['date'], inplace=True)

# List of columns to sum during aggregation (excluding 'date')
columns_to_sum = df_sum.columns.drop('date').tolist()

# Group by the 'date' column and sum the specified columns
df_aggregated_sum = df_sum.groupby('date')[columns_to_sum].sum().reset_index()

# Sort the DataFrame by the 'date' column in ascending order
df_aggregated_sum = df_aggregated_sum.sort_values(by='date', ascending=True)

# Format the 'date' column back to 'ngày/tháng/năm' (DD/MM/YYYY) as requested
df_aggregated_sum['date'] = df_aggregated_sum['date'].dt.strftime('%d/%m/%Y')

# Display the resulting aggregated and sorted DataFrame
print(df_aggregated_sum)

# Optionally, save the result to a new CSV file
df_aggregated_sum.to_csv('Hashtag_sum_aggregated_sorted_8_9_10.csv', index=False)
```
